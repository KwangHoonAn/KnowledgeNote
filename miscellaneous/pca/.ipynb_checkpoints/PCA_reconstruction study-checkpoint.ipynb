```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('/tmp/tensorflow/mnist/input_data', one_hot=False)
```

```python
x_, y_ = mnist.train.next_batch(10000)
x_.shape
# x_ = np.reshape(x, [28, 28])
# x_.shape
```

```python
y_[:10]
```

```python
X = [x for x, y in zip(x_, y_) if y == 3]
len(X)
```

```python
import numpy as np
mu = np.mean(X, axis=0)
X_center = X - mu
np.sum(X_center, axis=0)[0]
# sum(X_center[0])
```

```python
(X_center - X).shape
# (X_center - X)
```

```python
# Cov [784, 784 dimension]
Cov = np.dot( (X_center - X).T, (X_center - X) )
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.imshow(mu.reshape([28, 28]), 'gray')
```

```python
plt.imshow(X[0].reshape([28, 28]), 'gray')
```

```python
eig_val, eig_vec = np.linalg.eig(Cov)
eig_val.shape, eig_vec.shape
```

```python
print(eig_val[:10])
```

```python
eig_val = eig_val.astype('float')
eig_vec = eig_vec.astype('float')
eig_val[:10]
```

```python
# maximum index
indices = np.argsort(eig_val)[::-1]

for i in indices[:3]:
    plt.imshow(eig_vec[:,i].reshape([28, 28]), 'gray')
    plt.xlabel('eigen_val %d' %(eig_val[i]))
    plt.show()
    print(eig_val[i])
```

```python
reconstructions = [1, 100, 500, 784]
for i in reconstructions:
    subset_vec = eig_vec[:,:i]
    # [200, 784] * [784]
    inner = np.dot(subset_vec.T, x_[0])
    x_k = np.dot(subset_vec, inner)
    print("reconstruction errors to mu : ", sum((x_k-x_[0])**2) )
#     print(x_k)
    plt.imshow(x_k.reshape([28,28]), 'gray')
    plt.show()
```
