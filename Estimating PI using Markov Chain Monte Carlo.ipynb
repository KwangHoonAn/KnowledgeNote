```python
import numpy as np
np.random.random_sample() - 0.5
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def random_generator():
    x = np.random.random_sample() - 0.5
    y = np.random.random_sample() - 0.5
    return x, y
```

```python
def inSideCircle(x, y):
    # x^2 + y^2 = radius^2
    raidus_square = (x**2) + (y**2)
    return raidus_square < (0.5)**2
```

```python
circlePoints = 0
squarePoints = 0.0
estimated_pi = 0
epoch = 1000
estimations = []
inner_x = []
inner_y = []
outer_x = []
outer_y = []
```

```python
for i in range(epoch):
    x_, y_ = random_generator()
    if inSideCircle(x_, y_):
        circlePoints += 1
        inner_x.append(x_)
        inner_y.append(y_)
    else:
        outer_x.append(x_)
        outer_y.append(y_)
    squarePoints += 1
    estimated_pi = 4*circlePoints / squarePoints
    estimations.append(estimated_pi)
```

```python
fig, ax = plt.subplots()
ax.plot(range(0, len(estimations), 1), estimations, label='pi value')
ax.set_xlabel('Epoch')
ax.set_ylabel('Numerical PI value')
ax.set_title('Estimated PI per epoch')
ax.legend(loc=0)
```

```python
plt.plot(outer_x, outer_y, 'o',
             label="outer='{0}'".format('o'))
plt.plot(inner_x, inner_y, 'x',
             label="inner='{0}'".format('x'))
circle = plt.Circle((0, 0), 0.5, color='r', fill=False)
ax = plt.gca()
ax.add_patch(circle)
ax.plot()
```

```python
### Explanation
### Above circle has a radius 0.5 enclosed by 1 x 1 square
### Area of the circle -> PI*Radius^2 = PI*(1/4) and square area = 1*1
### Circle Area / Square area, then PI*(1/4) / 1 = PI*(1/4)
### This ratio represents ratio of randomly sampled datas from c
```
